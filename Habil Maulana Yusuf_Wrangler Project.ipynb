## 1.6 Adjust Salary to Present Values

### Inflation Rate

```python
# Inflation rates
us_inflation_rates = {2019: 0.0181,
                      2020: 0.0123,
                      2021: 0.0470,
                      2022: 0.065}
global_inflation_rates = {2019: 0.0219,
                          2020: 0.0192,
                          2021: 0.0350,
                          2022: 0.088}

def adjust_salary(x):
  year = x['work_year']
  original_salary = x['salary_in_usd']
  currency = x['salary_currency']

  # Kondisi pertama (Kalau 2023, maka tidak kena inflasi)
  if year == 2023:
    return original_salary
  else:
    adjust_salary = original_salary
    for y in range(year, 2023): # Loop sampai tahun 2022
      # kondisi kedua (Kalau currency adalah USD gunakan us inflation rates, selain USD gunakan global)
      if currency == 'USD':
        inflation_rate = us_inflation_rates[y]
      else:
        inflation_rate = global_inflation_rates[y]
      adjust_salary = adjust_salary * (1 + inflation_rate)

    return adjust_salary
```

```python
df_c['adjusted_salary'] = df_c.apply(adjust_salary, axis=1)
```
